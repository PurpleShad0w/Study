```python
import numpy as np
from pandas._libs.missing import NA
import pandas as pd
import random
import missingno as msno
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('url.csv')
```

```python
pages = []
for i in range(len(df)):
    if random.randint(1,6) == 1:
        pages.append(NA)
    else:
        pages.append(np.random.normal(loc=200, scale=50))
df['pages'] = pages
df.head(10)
```

```python
msno.matrix(df[['pages']])
```
